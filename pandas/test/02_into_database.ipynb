```python
import pandas as pd
import pymysql
```

```python
file_path = r'../source/company_info.csv'
df = pd.read_csv(file_path, skipinitialspace=True)
df
```

```python
config = {
    "host": "localhost",
    "port": 3306,
    "user": "root",
    "password": "password",
    "db": "EXAMPLE",
    "charset":  "utf8mb4",
}
```

```python
conn = pymysql.connect(**config)
cursor = conn.cursor()
sql_insert = """
    insert into company_info
    values (20241103, 1101, '臺灣水泥股份有限公司', 1);
"""
cursor.execute(sql_insert)
conn.commit()

cursor.close()
conn.close()
```

```python
conn = pymysql.connect(**config)
cursor = conn.cursor()

sql_insert_many = """
insert into company_info
values (%s, %s, %s, %s);
"""

values = list(df.itertuples(index=False, name=None))

cursor.executemany(sql_insert_many, values)
conn.commit()

cursor.close()
conn.close()
```
