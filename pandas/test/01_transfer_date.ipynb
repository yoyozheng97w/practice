```python
import pandas as pd
```

```python
file_path = r'../source/date_transfer.csv'
df = pd.read_csv(file_path, skipinitialspace=True)
df
```

```python

for i in range(len(df)):
    df.loc[i, 'date'] = str(df.loc[i, 'date'])
    df.loc[i, 'date'] = str(int(df.loc[i, 'date'][0:3]) + 1911) + df.loc[i, 'date'][3:]
df
```
