```python
import pandas as pd
```

```python
file_path = r'../source/raw.csv'
df = pd.read_csv(file_path, skipinitialspace=True)
df
```

```python
condition = df['price'] < 0
df = df.drop(df[condition].index)
df
```

```python
# 迴圈寫法
for idx in df.index:
    if df.loc[idx, 'price'] < 0:
        df.drop(idx, inplace=True)
df
```
